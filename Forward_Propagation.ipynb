```python
def forward_chaining(rules, facts, goal):
    print("--- Starting Forward Chaining ---")
    print(f"Initial Facts: {facts}\n")

    new_fact_added = True
    cycle = 1
    while new_fact_added:

        new_fact_added = False
        print(f"Cycle {cycle}:")

        for rule in rules:

            if rule['triggered']:
                continue

            if all(cond in facts for cond in rule['if']):

                result = rule['then']

                if result not in facts:
                    facts.append(result)
                    new_fact_added = True
                    rule['triggered'] = True
                    print(f"  > Rule Fired: IF {' AND '.join(rule['if'])} THEN {result}")

        print(f"  Facts at end of cycle {cycle}: {facts}")

        if goal in facts:
            print(f"\nGoal '{goal}' has been PROVEN!")
            return True

        cycle += 1
        if not new_fact_added:
            print("\nNo more rules can be fired.")

    if goal not in facts:
        print(f"\nGoal '{goal}' could not be proven.")
    return False

rules_list = [
    {'if': ['shape is long', 'color is yellow'], 'then': 'fruit is banana', 'triggered': False},
    {'if': ['shape is round', 'color is red', 'size is medium'], 'then': 'fruit is apple', 'triggered': False},
    {'if': ['shape is round', 'color is red', 'size is small'], 'then': 'fruit is cherry', 'triggered': False},
    {'if': ['skin_smell'], 'then': 'perfumed', 'triggered': False},
    {'if': ['fruit is lemon'], 'then': 'citrus_fruit', 'triggered': False},
    {'if': ['fruit is orange'], 'then': 'citrus_fruit', 'triggered': False},
    {'if': ['fruit is pomelo'], 'then': 'citrus_fruit', 'triggered': False},
    {'if': ['fruit is grapefruit'], 'then': 'citrus_fruit', 'triggered': False},
    {'if': ['size is medium', 'color is yellow', 'perfumed'], 'then': 'fruit is lemon', 'triggered': False},
    {'if': ['size is medium', 'color is green'], 'then': 'fruit is kiwi', 'triggered': False},
    {'if': ['size is big', 'perfumed', 'color is orange', 'citrus_fruit'], 'then': 'fruit is grapefruit', 'triggered': False},
    {'if': ['perfumed', 'color is orange', 'size is medium'], 'then': 'fruit is orange', 'triggered': False},
    {'if': ['perfumed', 'color is red', 'size is small', 'seeds = 0'], 'then': 'fruit is strawberry', 'triggered': False},
    {'if': ['diameter = 7'], 'then': 'size is medium', 'triggered': False}
]

initial_facts = ['seeds = 0', 'diameter = 7', 'skin_smell', 'color is orange']
goal_to_prove = 'citrus_fruit'

forward_chaining(rules_list, initial_facts, goal_to_prove)
```
